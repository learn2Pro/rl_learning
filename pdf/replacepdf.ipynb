```python
import fitz  # PyMuPDF

def replace_text_in_pdf(input_path, output_path, old_text, new_text, custom_font):
    # 打开 PDF 文件
    doc = fitz.open(input_path)
    # 遍历所有页面
    for page in doc:
        # 搜索目标文本
        text_instances = page.search_for(old_text)
        for inst in text_instances:
            # 覆盖原文本
            rect = fitz.Rect(inst)
            page.draw_rect(rect, color=(1, 1, 1), fill=(1, 1, 1))  # 用白色填充

            # 确保文本精确对齐到原始位置
            x, y = rect.top_left  # 获取左上角坐标
            page.insert_text(
                (x, y+8),           # 文本起始位置
                new_text,         # 新文本
                fontfile=custom_font,
                fontsize=10,      # 调整为合适的字体大小
                color=(0, 0, 0),  # 黑色
            )

    # 保存结果
    doc.save(output_path)
    doc.close()

    print(f"文本已替换并保存到 {output_path}")
    
# 文件路径
input_path = "/Users/bytedance/Downloads/Receipt-2914-1570-8386 (1).pdf"
output_path = "Invoice-E2E0C6C1-0008 (1)-replaced.pdf"
output_path1 = "Invoice-E2E0C6C1-0008 (1)-replaced1.pdf"
output_path2 = "Invoice-E2E0C6C1-0008 (1)-replaced2.pdf"

# 需要替换的文本
old_text = "DERONG TANG"
new_text = "YAN XING"
# 自定义字体
custom_font = "InterDisplay-Thin.ttf"

replace_text_in_pdf(input_path, output_path, "唐云", new_text, custom_font)
replace_text_in_pdf(output_path, output_path1, "DERONG TANG", new_text, custom_font)
replace_text_in_pdf(output_path1, output_path2, "druidlangde@gmail.com", "yanxing@buaa.edu.cn", custom_font)
```

```python
import fitz

def extract_text_colors(pdf_path):
    """
    从 PDF 中提取文本及其字体颜色。
    """
    # 打开 PDF 文件
    doc = fitz.open(pdf_path)

    # 遍历每一页
    for page_num in range(len(doc)):
        page = doc[page_num]

        # 获取页面内容作为字典
        text_dict = page.get_text("dict")

        # 遍历块，提取颜色信息
        for block in text_dict["blocks"]:
            print(block.keys())
            if "lines" not in block:
                continue
            for line in block["lines"]:
                for span in line["spans"]:
                    text = span["text"]  # 文本内容
                    color = span["color"]  # 文本颜色（RGB值，整数表示）

                    # 转换颜色值（十进制整数转为RGB元组）
                    r = (color >> 16) & 0xFF
                    g = (color >> 8) & 0xFF
                    b = color & 0xFF

                    print(f"Page {page_num + 1}: Text: '{text}', Color: RGB({r}, {g}, {b})")

    # 关闭 PDF
    doc.close()

# 示例调用
extract_text_colors("/Users/bytedance/Downloads/Invoice-E2E0C6C1-0008 (1).pdf")
```

```python
import fitz  # PyMuPDF

# 原始PDF文件路径
input_pdf_path = "/Users/bytedance/Downloads/Invoice-E2E0C6C1-0008 (1).pdf"
# 输出PDF文件路径
output_pdf_path = "Modified_Receipt.pdf"

# 修改PDF中的文本
def modify_pdf(input_path, output_path):
    # 打开PDF文件
    doc = fitz.open(input_path)

    # 遍历每一页
    for page_num in range(len(doc)):
        page = doc.load_page(page_num)
        text_instances = page.search_for("DERONG TANG")  # 查找要替换的文本

        # 如果找到目标文本
        if text_instances:
            for inst in text_instances:
                # 在目标位置添加一个白色矩形覆盖原文本
                page.add_redact_annot(inst, fill=(1, 1, 1))  # 白色填充
                page.apply_redactions()

                # 在相同位置写入新文本
                rect = fitz.Rect(inst)  # 获取原文本的位置
                page.insert_text(
                    rect.top_left,  # 文本插入位置
                    "YAN XING",  # 新文本
                    fontsize=12,  # 字体大小（根据实际情况调整）
                    fontname="helv",  # 字体名称
                    color=(0, 0, 0),  # 文本颜色（黑色）
                )

    # 保存修改后的PDF
    doc.save(output_path)
    print(f"Modified PDF saved to: {output_path}")

modify_pdf(input_pdf_path, output_pdf_path)
```
